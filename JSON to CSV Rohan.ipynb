```python
import os
import sys
import csv
import pandas as pd
import json
```

```python
data = pd.read_json("classified_profiles.json", dtype=False)
data = data.T
```

```python
data = data.iloc[:,[15,7,9,3,12,19,10,11,25,6,0]]
```

```python
data = data.reset_index(drop=True)
data
```

```python
def calculate_length(x):
    return len(x)

def fake(x):
    if x:
        return 0
    else:
        return 1

def is_private(x):
    if x:
        return 1
    else:
        return 0

def ext_url_present(x):
    if x:
        return 1
    else:
        return 0

def anonymous_profile_picture(x):
    if x:
        return 1
    else:
        return 0
```

```python
username_length = (data['username'].apply(lambda x: calculate_length(x)))
```

```python
data['username_length'] = username_length
```

```python
bio_length = (data['biography'].apply(lambda x: calculate_length(x)))
data['bio_length'] = bio_length
```

```python
full_name_len = (data['full_name'].apply(lambda x: calculate_length(x)))
data['full_name_len'] = full_name_len
```

```python
fake = (data['is_authentic'].apply(lambda x: fake(x)))
data['fake'] = fake
```

```python
is_private = (data['is_private'].apply(lambda x: is_private(x)))
data['is_private'] = is_private
```

```python
external_url = (data['external_url'].apply(lambda x: ext_url_present(x)))
data['ext_url_present'] = external_url
```

```python
profile_pic = (data['has_anonymous_profile_picture'].apply(lambda x: anonymous_profile_picture(x)))
data['has_anonymous_profile_picture'] = profile_pic
```
